```python
#use this library to make a decision between left or right movement
## also assume a set of steps to walk in each direction, lets say steps= [1,2,3,4,5,6], while moving to
# each direction one also has to incorporate number of steps
from random import choice

class RandomWalk():
    """A class to generate random walks."""
    def __init__(self, no_of_step=10000):
        """Initialize attributes of a walk."""
        self.no_of_step = no_of_step
        # All walks start at (0, 0).
        self.x = [0]*no_of_step
        self.y = [0]*no_of_step
    def fill_walk(self):
        """Calculate all the points in the walk."""
        for i in range(1, self.no_of_step):
            val = choice([1, 2, 3, 4])
            if val == 1:
                self.x[i] = self.x[i - 1] + 1
                self.y[i] = self.y[i - 1]
            elif val == 2:
                self.x[i] = self.x[i - 1] - 1
                self.y[i] = self.y[i - 1]
            elif val == 3:
                self.x[i] = self.x[i - 1]
                self.y[i] = self.y[i - 1] + 1
            else:
                self.x[i] = self.x[i - 1]
                self.y[i] = self.y[i - 1] - 1
       
```

```python
import matplotlib.pyplot as plt
# Make a random walk, and plot the points.
rw = RandomWalk()
rw.fill_walk()
plt.figure(figsize=(10,8))
plt.scatter(rw.x, rw.y, s=15)
plt.show()
```
