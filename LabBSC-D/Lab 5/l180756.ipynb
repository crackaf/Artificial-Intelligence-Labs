```python
max_hours = data['hours-per-week'].max()
print("Max time is {0} hours per week.".format(max_hours))

max_workers = len(data[data['hours-per-week'] == max_hours].index)

print("Total number of such hard workers {0}".format(max_workers))

rich_earning = float(len(data[(data['hours-per-week'] == max_hours)
                 & (data['salary'] == '>50K')].index)) / max_workers
print("Percentage of rich among them {0}%".format(int(100 * rich_earning)))
```
